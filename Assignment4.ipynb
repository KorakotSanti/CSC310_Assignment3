```python
%matplotlib inline
import pandas as pd
from sklearn import tree
from assets.treeviz import tree_print
from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
epilepsy_df = pd.read_csv('./assets/epilepsy.csv')
```

```python
epilepsy_df.head()
```

```python
# dropped Unamed: 0 because it is a index
# dropped ID because it is not relevant to the result
# dropped Age10 and Base4 because it is pretty much a duplicate of Age and Base column
epilepsy_df = epilepsy_df.drop(['Unnamed: 0','ID','Age10','Base4'], axis=1)
```

```python
epilepsy_target = pd.DataFrame(epilepsy_df['Trt'])
epilepsy_features = epilepsy_df.drop('Trt', axis=1)
```

```python
sns.pairplot(epilepsy_df,hue='Trt')
```

```python
epilepsy_df.loc[:,'Trt'].value_counts().plot.bar()
```

All of our independent variables except for age is unimodal. Our independent variable age is multimodal. The independent variables that are unimodal are also positively skewed. We can identified clusters between variables in our scatter plots except for the scatter where age is one of the variable. It is a bit hard to tell, but most of the progabide (treatment) cluster are all in the lower end of the scatterplot with the exception of the tiny outliers. While the placebo result cluster are overlapped a little over the progabide cluster, while at the same time remained in a more higher position in the plot. There seems to be no cluster between age and any other variables in our model. The clusters are related to our decision tree model from last lab. Our data is balanced.

```python
dtree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=6)
```

```python
dtree.fit(epilepsy_features,epilepsy_target)
```

```python
tree_print(dtree,epilepsy_features)
```

```python
epi_predict = dtree.predict(epilepsy_features)
epi_predict_df = pd.DataFrame(epi_predict)
epi_predict_df.columns = ['Trt']
```

```python
epi_predict_df.head()
```

```python
epilepsy_target.head()
```

```python
print(f"The accuracy is {accuracy_score(epilepsy_target,epi_predict_df)}")
```

This is the closest to at least 80% of our best model. Our new decision tree models does not differ too much from our old model in terms of interpretability.
